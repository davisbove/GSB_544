```python
import pandas as pd
from palmerpenguins import load_penguins
from plotnine import ggplot, geom_point, aes, geom_boxplot
import plotnine as p9
```

```python
penguins = load_penguins()

(ggplot(penguins, aes(x = "species", y = "bill_length_mm", fill = "species"))
+ geom_boxplot()
)
```

```python
penguins.head()
```

Can also change the scale of the graph
  changing the element included in the graph

```python
(ggplot(penguins, aes(x= "species", y = "bill_depth_mm", fill = "species")) + geom_boxplot()
+p9.scale_y_reverse())
```

```python
(ggplot(penguins, aes(x = "species", y = "bill_length_mm", fill = "island")) + geom_boxplot())
```

```python

(ggplot(penguins, aes(x = "species", y = "bill_length_mm", fill = "species"))
+ geom_boxplot(fill = ("cornflowerblue", "green", "red")
)
+ p9.theme_classic())
```

Do not include bar plots for numeric data
use bar plots for categorical counts only

```python
from plotnine import geom_bar
(ggplot(penguins, aes( x = "species"))
+ geom_bar()
)
```

```python
(ggplot(penguins,
aes(
  x = "species",
  fill = "island"
))
+ geom_bar(position = "dodge")
)
```

```python
(ggplot(penguins,
aes(
  x = "species",
  y = "bill_length_mm",
  fill = "sex"
))
+ geom_boxplot()
)
```

Histograms - statisticians love histograms
    shows just one variable which is a numeric variable
    can convert to a categorical and shows distribution


```python
from plotnine import geom_histogram
(ggplot(penguins,
aes(
  x = "bill_length_mm"
))
+ geom_histogram()
)
```

in histograms you can chose the size of bins, or the number of total bins to make


```python
(ggplot(penguins,
aes(
  x = "bill_length_mm"
))
+ geom_histogram(bins = 10)
)
```

The next step to a histogram is a density, you can see distributions in a single plot

    can also include the alpha command to control the "see-throughness"

```python
from plotnine import geom_density
(ggplot(penguins,
aes(
  x = "bill_length_mm",
  fill = "species"
))
+ geom_density(alpha = 0.5)
)
```

Scatterplot is denoted as geom_point()

```python
(ggplot(penguins, aes(x = "bill_depth_mm", y = "bill_length_mm", color = "species")) + geom_point())
```

facet wrapping
    makes side by side plots of the same data

```python
from plotnine import facet_wrap
(ggplot(penguins,
aes(
  x = "species",
  y = "bill_length_mm"
))
+ geom_boxplot()
+ facet_wrap("sex")
)
```
